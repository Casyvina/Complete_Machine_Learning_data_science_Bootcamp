```python
# Let's # import our packages
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Let's import our "car_sales.csv" file

car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
# Let's add a date_range column
car_sales["Sale Date"] = pd.date_range("6/26/2023", periods=len(car_sales))
car_sales
```

```python
car_sales["Price"] = car_sales["Price"].str.replace("[\$\,\.]", '', regex=True).str[:-2].astype(int) 
car_sales
```

```python
type(car_sales["Price"]), dtype(car_sales["Price"])
```

```python
# Let's create a "Total Sale" with cumcum
car_sales["Total Sale"] = car_sales["Price"].cumsum()
car_sales
```

```python
# Let's plot a Graph for Total Sale against Sale date
car_sales.plot(x ="Sale Date", y ="Total Sale");
```

```python
# Let's create a scatter plot
car_sales.plot(x="Odometer (KM)", y="Price", kind="scatter");
```

```python
# Ploting a Time Series
ts = pd.Series(np.random.randn(1000),
               index =pd.date_range("6/26/2023", periods=1000))
ts = ts.cumsum()
ts.plot();
```

```python
# let's import our raw dataset
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease.head()
```

```python
# Let's create a OO methods
# There are two types of Matplotlib 
# 1. Pyplot method
# 2. Ojected Oriented method

heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
# Pyplot method
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             figsize=(10, 6));
```

```python
# Pyplot mixed with OOP
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             ax=ax)
ax.set_xlim([45, 80]);
```

```python
over_50.target.values
```

```python
heart_disease.head()
```

```python
# OOP from scratch
# first create subplot and fig
fig, ax = plt.subplots(figsize=(10, 6))

# plot the data 
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

# Customized the plot
ax.set(xlabel="Age",
       ylabel="Cholesterol",
       title="Heart Disease and Cholesterol Level")

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target")

# Let's draw a line
ax.axhline(over_50["chol"].mean(), 
           linestyle = "-");
```

```python
# Now let's try to practise with subplots
# First create and fig, ax 
fig, (ax0, ax1, ax2) = plt.subplots(nrows=3,
                               ncols=1,
                               figsize=(10, 10),
                               sharex=True)
# Add our data with scatter plot
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"]);

# Customized out data
ax0.set(title="Heart Disease and Cholesterol Levels",
        ylabel="Cholesterol",
       )
# Add a legend tag
ax0.legend(*scatter.legend_elements(), title="Target")
# Add a meanline
ax0.axhline(over_50["chol"].mean(), 
            linestyle='--');

# Add our data with bar plot
bar = ax1.bar(x=over_50["age"],
              height= over_50["thalach"]);
# Customized our data
ax1.set(title="Heart Disease and Max Heart Rate",
        xlabel="Ages",
        ylabel="Max Heart Rate");
# Add a legend
# ax1.legend(*bar.legend_elements(), title="target") has non attribute BarContainer

# Draw a meanline
ax1.axhline(over_50["thalach"].mean(), 
            linestyle='--');

# Let's create one more with hist
hist = ax2.hist(heart_disease["age"], bins=20)

# customize our data
ax2.set(title="Heart Disease Ages distribution",
        xlabel="Ages",
        ylabel="Range");
# No legend am pretty sure
# Add a meanline
ax2.axhline(heart_disease["age"].mean(),
            linestyle='--');

# Add a general title
fig.suptitle(t="Heart Disease", fontsize=16, fontweight="bold");
```
